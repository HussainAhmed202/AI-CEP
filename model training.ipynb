# Handling Missing Values

```python
# percentage of missing data in each column
data.isnull().sum()/data.shape[0]*100
```

```python
# filling feature with large percentage of missing data using random choice 
lst=['Evaporation','Sunshine','Cloud9am','Cloud3pm']
for col in lst:
    fill_list = data[col].dropna()
    data[col] = data[col].fillna(pd.Series(
        np.random.choice(fill_list , size = len(data.index))))
#categorical features
s = (data.dtypes == "object")
object_cols = list(s[s].index)
```

```python
for i in object_cols:
    data[i].fillna(data[i].mode()[0], inplace=True)
```

```python
#numeric features
t = (data.dtypes == "float64")
num_cols = list(t[t].index)
```

```python
for i in num_cols:
    data[i].fillna(data[i].median(), inplace=True)
```

# Encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
#Encoding 
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
for i in object_cols:
    data[i] = label_encoder.fit_transform(data[i])
```

```python
X = data.drop(['RainTomorrow','day'],axis=1).values
y = data['RainTomorrow'].values
```

# Modeling

```python
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25,random_state=101)
```

```python
scaler = MinMaxScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation,Dropout
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.layers import Dropout

from keras import regularizers
```

```python

model = Sequential([
    Dense(units=24,activation='tanh',),
    Dense(units=18,activation='tanh',kernel_regularizer=regularizers.l2(0.0001)),
    Dense(units=23,activation='tanh',kernel_regularizer=regularizers.l2(0.003)),
    Dropout(0.5),
    Dense(units=12,activation='tanh'),
    Dropout(0.2),
    Dense(units=1,activation='sigmoid')
])
```

```python
# For a binary classification problem
opt = Adam(learning_rate=0.001)
model.compile(loss='binary_crossentropy', optimizer=opt)
```

```python
early_stop = EarlyStopping(monitor='val_loss', mode='min',min_delta=0.001, verbose=1, patience=5,restore_best_weights=True)
```

```python
model.fit(x=X_train, 
          y=y_train, 
          validation_data=(X_test, y_test), verbose=1,
          callbacks=[early_stop],
          epochs = 150,
          batch_size = 32
)
```

# Model Evaluation

```python
from sklearn.metrics import classification_report,confusion_matrix
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error , mean_absolute_percentage_error , mean_squared_error
```

```python
model_loss = pd.DataFrame(model.history.history)
model_loss.plot()
```

```python
predictions = model.predict(X_test)
predictions=(predictions>0.5)
```

```python
confusion_matrix(y_test,predictions)
```

```python
print(classification_report(y_test,predictions))
```

```python
mean_squared_error(y_test ,predictions)
```

```python
mean_absolute_error(y_test , predictions)
```

```python
r2_score(y_test , predictions)
```
